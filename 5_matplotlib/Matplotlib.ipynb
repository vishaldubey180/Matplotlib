## Histogram

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bloodsugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
```

#### 1)How many patient are normal
#### 2)how many are pre-diabetic
#### 3)how many diabetic

```python
plt.hist(bloodsugar)
```

```python
## bins and bucket in histogram
```

```python
plt.hist(bloodsugar, bins=3)
```

```python
plt.hist(bloodsugar, bins=3, rwidth=0.95)
```

```python
plt.hist(bloodsugar, bins=[80,100,125,150], rwidth=0.95)
```

```python
plt.hist(bloodsugar, bins=[80,100,125,150], rwidth=0.95, color='g')
```

```python
plt.hist(bloodsugar, bins=[80,100,125,150], rwidth=0.95, color='g', histtype='step')
```

```python
blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]
```

```python
 plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['g', 'r'],
               label=['men', 'Women'])
plt.legend()
```

```python
plt.xlabel("sugar range")
plt.ylabel("Total Number of patients")
plt.title('blood sugar analysis')
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['g', 'r'],
               label=['men', 'Women'])
plt.legend()
```

```python
# we can change orientation also
```

```python
plt.xlabel("sugar range")
plt.ylabel("Total Number of patients")
plt.title('blood sugar analysis')
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['g', 'r'],
               label=['men', 'Women'], orientation='horizontal')
plt.legend()
```
