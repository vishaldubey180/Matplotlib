```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# lwt's create weather data
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41] 
```

```python
plt.plot(days,max_t)
```

```python
# we can plot all charts in one graph
```

```python
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
## xlabel and y label are misisng
```

```python
plt.xlabel('Weekdays')
plt.ylabel('temperature')
plt.title("WEATHER")
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

## create legend

```python
# first give label to each properties
```

```python
plt.xlabel('Weekdays')
plt.ylabel('temperature')
plt.title("WEATHER")
plt.plot(days,max_t, label='max')
plt.plot(days,min_t, label='min')
plt.plot(days,avg_t, label='avg')
```

```python
# set legend
```

```python
plt.xlabel('Weekdays')
plt.ylabel('temperature')
plt.title("WEATHER")
plt.plot(days,max_t, label='max')
plt.plot(days,min_t, label='min')
plt.plot(days,avg_t, label='avg')

plt.legend()
```

```python
# we can change location of label also
```

```python
plt.xlabel('Weekdays')
plt.ylabel('temperature')
plt.title("WEATHER")
plt.plot(days,max_t, label='max')
plt.plot(days,min_t, label='min')
plt.plot(days,avg_t, label='avg')

plt.legend(loc='upper right')
```
