```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
plt.plot(x,y)
```

 # formate string

```python
plt.plot(x,y, 'g+') # g+ green marker
```

```python
plt.plot(x,y, 'g+--')
```

```python
plt.plot(x,y, 'g+--')
```

```python
# order doesen't matter
```

## Individual argument for plot

```python
plt.plot(x,y, color='red', marker='D', linestyle='')
```

```python
plt.plot(x,y, color='red', marker='D', linestyle='', markersize=20)
```
