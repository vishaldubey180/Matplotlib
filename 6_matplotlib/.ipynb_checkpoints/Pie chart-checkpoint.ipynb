## Pie chart

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ['home rant', 'Food', 'phone/internate bill', 'Car', 'Other utilities']
```

```python
plt.pie(exp_vals, labels=exp_labels)
```

```python
plt.pie(exp_vals, labels=exp_labels)
plt.axis('equal')
```

```python
# to remove extra labels
```

```python
plt.pie(exp_vals, labels=exp_labels)
plt.axis('equal')
plt.show()
```

```python
# we can also increase radius of our pie chart
```

```python
plt.pie(exp_vals, labels=exp_labels, radius=2)
plt.show()
```

```python
# to find percentage
```

```python
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.1f%%')
plt.show()
```

```python
# use of explode
```

```python
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.1f%%', explode = [0,0,0,0,0])
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.1f%%', explode = [0,0.2,0,0,0])
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.1f%%', explode = [0,0.8,0,0,0])
plt.show()
```

```python
# we can start our piechart at different angle also
```

```python
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.1f%%', startangle=45)
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.1f%%', startangle=120)
plt.show()
```
